```python
movies = movies.merge(credits, on = "title")
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast', 'crew']]
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
def preprocess_genres(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i["name"])
    return L
```

```python
movies['genres'] = movies['genres'].apply(preprocess_genres)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(preprocess_genres)
```

```python
movies.head()
```

```python
def preprocess_cast(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i["name"])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(preprocess_cast)
```

```python
def preprocess_crew(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i["name"])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(preprocess_crew)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x : x.split())
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies['tags'].apply(lambda x : " ".join(x))
```

```python
movies_final_data = movies[['movie_id', 'title', 'tags']]
```

```python
movies_final_data['tags']
```

```python
movies_final_data['tags'] = movies_final_data['tags'].apply(lambda x : " ".join(x))
```

```python
movies_final_data.head()
```

```python
movies_final_data['tags'] = movies_final_data['tags'].apply(lambda x : x.lower())
```

```python
movies_final_data.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(movies_final_data['tags']).toarray()
```

```python
cv.get_feature_names()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y = []
  for i in text.split() : 
    y.append(ps.stem(i))

  return " ".join(y)
```

```python
movies_final_data['tags'] = movies_final_data['tags'].apply(stem)
```

```python
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(movies_final_data['tags']).toarray()
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape
```

```python
similarity = cosine_similarity(vectors)
similarity[0]
```

```python
def recommend_movie(movie):
  movie_index = movies_final_data[movies_final_data['title'] == movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)), reverse = True, key = lambda x : x[1])[1:6]

  for i in movies_list :
    print(movies_final_data.iloc[i[0]].title)
```

```python
recommend_movie('Avatar')
```
